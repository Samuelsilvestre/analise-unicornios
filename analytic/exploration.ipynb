```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
PATH = os.path.join('.\\..\\data_raw\\unicorns till sep 2022.csv')

df = pd.read_csv(PATH)
```

```python
# os primeiros 10 registros
df.head(10)
```

```python
# colunas e tipo de dados

df.columns

df.dtypes
```

```python
# analise grafica dos campos nulos

plt.figure(figsize=(18,10))
plt.title('mapa mapeando os campos nulos')
sns.heatmap(df.isnull())
```

```python
df = df.dropna()
plt.figure(figsize=(12,6))
plt.title('mapa mapeando os campos nulos')
sns.heatmap(df.isnull())
```

```python
df['Valuation ($B)']
```

```python
df['Date Joined']
```

```python
# tratamento da data
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
```

```python
# tratamento do valor
df['Valuation ($B)'] = df['Valuation ($B)'].apply(lambda x: float(x.split(sep='$')[1]))
```

```python
# salvando os dados tratados

df.head()

PATH_SAVE = os.path.join('.//..//data-ready//dataset_unicornios.csv')
df.to_csv(PATH_SAVE, sep=',', index=False)
```
