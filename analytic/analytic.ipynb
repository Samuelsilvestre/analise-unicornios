```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
PATH = os.path.join('.\\..\\data-ready\\\dataset_europa_unicornios.csv')
df = pd.read_csv(PATH)
```

```python
analise_pais = df.groupby( by=['Country'] ).sum()['Valuation ($B)'].reset_index()
analise_pais.sort_values('Valuation ($B)', ascending=False)

```

```python
analise_setor = df['Industry'].value_counts(normalize=True) *100

plt.figure(figsize=(18,12))
plt.title('grafico dos setores na europa')
plt.pie(
    analise_setor,
    labels=analise_setor.index,
    shadow=True,
    autopct='%1.1f%%'
)
plt.show()
```

```python
percentual_pais = df['Country'].value_counts(normalize=True) * 100

plt.figure(figsize=(20,12))
plt.title('grafico paises geradores de unicornios europa')
plt.pie(
    percentual_pais,
    labels=percentual_pais.index,
    shadow=True,
    autopct="%1.1ff%%"
)
plt.show()
```
