```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os

PATH = os.path.join('.\\..\\data-ready\\dataset_unicornios.csv')
df = pd.read_csv(PATH)
```

```python
df['Country'].head(20)
```

```python
# analise do percentual dos paises
percentual_pais = df['Country'].value_counts(normalize=True) * 100
percentual_pais
```

```python
plt.title('grafico de barras de paises geradores de unicornios')
plt.figure(figsize=(20,12))
plt.bar(df['Country'].value_counts().index, df['Country'].value_counts())
plt.xticks(rotation=45)
plt.show()
```

```python
# grafico de pizza
plt.figure(figsize=(18,12))
plt.title('percentual de paises geradores de unicornios')
plt.pie(
    percentual_pais.head(20),
    labels=percentual_pais.index[0:20],
    shadow=True,
    autopct='%1.1f%%'
)
```

```python
# gerando um coluna ano
df['Year'] = pd.DatetimeIndex(df['Date Joined']).year
```
